# Pengaturan Figure Layouts dengan GridSpec

## Import modules

```python
%matplotlib inline
```

```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import numpy as np
```

## Simple 2x2 layout dengan subplots

```python
siswa = ['Cecep', 'Joko', 'Wati', 'Ani']
uas = [68, 83, 96, 72]

fig, axes = plt.subplots(ncols=2, nrows=2, constrained_layout=True)

axes[0][0].bar(siswa, uas)
axes[0][1].pie(uas, labels=siswa)
axes[1][0].pie(uas, labels=siswa, explode=[0, 0, 0, 0.2])
axes[1][1].barh(siswa, uas)

plt.show()
```

Variable `fig` akan menampung objek figure, dan `axes` akan menampung objek axes. Karena akan dibuat plotting dengan 2x2 layout, maka parameter `ncols` dan `nrows` masing-masing diberi nilai 2. 

Parameter `constrained_layout` sendiri akan berhubungan dengan jarak dari masing-masing axesnya. Jika diberi nilai False, maka tiap axes-nya akan berdempetan.

## Simple 2x2 layout dengan GridSpec

```python
fig = plt.figure(constrained_layout=True) # membuat objek figure.

gs = gridspec.GridSpec(ncols=2, nrows=2, figure=fig) # membuat objek GridSpec.

ax1 = fig.add_subplot(gs[0, 0])
ax2 = fig.add_subplot(gs[0, 1])
ax3 = fig.add_subplot(gs[1, 0])
ax4 = fig.add_subplot(gs[1, 1])

ax1.bar(siswa, uas)
ax2.pie(uas, labels=siswa)
ax3.pie(uas, labels=siswa, explode=[0, 0, 0, 0.2])
ax4.barh(siswa, uas)

plt.show()
```

Prosesnya cukup sama seperti saat menggunakan subplots. Pada cara di atas, objek figure dibuat terlebih dahulu dengan menggunakan method `plt.figure` dan dilanjutkan dengan membuat objek GridSpec dengan menggunakan method `gridspec.GridSpec`.

Setelahnya barulah dibuat empat buah variable axes dan lanjut melakukan plotting sesuai kebutuhan.

```python
fig = plt.figure(constrained_layout=True) # membuat objek figure.

gs = fig.add_gridspec(ncols=2, nrows=2) # membuat objek GridSpec dengan method fig.add_gridspec

ax1 = fig.add_subplot(gs[0, 0])
ax2 = fig.add_subplot(gs[0, 1])
ax3 = fig.add_subplot(gs[1, 0])
ax4 = fig.add_subplot(gs[1, 1])

ax1.bar(siswa, uas)
ax2.pie(uas, labels=siswa)
ax3.pie(uas, labels=siswa, explode=[0, 0, 0, 0.2])
ax4.barh(siswa, uas)

plt.show()
```

Pada kode di atas, objek GridSpec dibuat dengan menggunakan fungsi `fig.add_gridspec`. Cara tersebut lebih efisien karena figurenya tidak perlu dispesifikasikan lagi.

## Rows Span dan Cols Span dengan GridSpec

```python
fig = plt.figure(constrained_layout=True) # membuat objek figure.

gs = fig.add_gridspec(ncols=2, nrows=2) # membuat objek GridSpec dengan method fig.add_gridspec

ax1 = fig.add_subplot(gs[0, :]) # menggunakan `:` untuk melakukan col span / perluasan kolom.
# ax2 = fig.add_subplot(gs[0, 1])
ax3 = fig.add_subplot(gs[1, 0])
ax4 = fig.add_subplot(gs[1, 1])

ax1.bar(siswa, uas)
# ax2.pie(uas, labels=siswa)
ax3.pie(uas, labels=siswa, explode=[0, 0, 0, 0.2])
ax4.barh(siswa, uas)

plt.show()
```

```python
fig = plt.figure(constrained_layout=True) # membuat objek figure.
gs = fig.add_gridspec(ncols=3, nrows=3) # membuat objek GridSpec dengan method fig.add_gridspec

ax1 = fig.add_subplot(gs[0, :])
ax1.set_title('gs[0, :]')

ax2 = fig.add_subplot(gs[1, :-1])
ax2.set_title('gs[0, -1]')
ax2.barh(siswa, uas)

ax3 = fig.add_subplot(gs[1:, -1]) # perluasan dari index baris 1 sampai baris index terakhir.
ax3.set_title('gs[1:, -1]')
ax3.bar(siswa, uas)

ax4 = fig.add_subplot(gs[-1, 0])
ax4.set_title('gs[-1, 0]')

ax5 = fig.add_subplot(gs[-1, -2])
ax5.set_title('gs[-1, -2]')

plt.show()
```

Kode di atas adalah contoh plotting dengan layout dasar 3x3, dimana terjadi beberapa perluasan kolom dan baris.
