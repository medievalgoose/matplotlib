## Simple Line Plot

```python
x = np.arange(0.0, 2.0, 0.01) #(starting point (inklusif), nilai akhir (eksklusif), interval)
x
```

```python
s = np.sin(2 * np.pi * x)
s
```

```python
fig, ax = plt.subplots() # OO style

ax.plot(x, s)
ax.set(xlabel="nilai x",
      ylabel="nilai y",
      title="Visualisasi nilai sin")

ax.grid()

plt.show()
```

```python
plt.plot(x,s) #pyplot style
plt.xlabel("nilai x")
plt.ylabel("nilai y")
plt.title("Visualisasi nilai sin")
plt.grid()
plt.show()
```

## Multiple Subplots

```python
x1 = np.linspace(0.0, 5.0, 100) #(min (inclusive), max(inclusive), banyaknya data point)
x2 = np.linspace(0.0, 2.0, 100)

y1 = np.cos(2 * np.pi * x1)
y2 = np.cos(2 * np.pi * x2)
```

```python
x1
```

### Multiple Subplots dengan OO style

```python
fig, (ax1, ax2) = plt.subplots(2, 1) #2 = jumlah baris, 1 = jumlah kolom.

fig.suptitle('Multiple Subplots')

ax1.plot(x1, y1, 'ro-')
ax1.set_ylabel('Nilai $cos(x_1)$') #Menggunakan Latex style.

ax2.plot(x2, y2, 'g.-')
ax2.set_ylabel('Nilai $cos(x_2)$')

ax2.set_xlabel('Nilai x')

plt.show()
```

### Multiple Subplots dengan pyplot Style

```python
plt.subplot(2, 1, 1) #Parameter ketiga untuk mengendalikan axes yang diinginkan.
plt.plot(x1, y1, 'ro-')
plt.title("Multiple Subplots")
plt.ylabel("Nilai $cos(x_1)$")

plt.subplot(2, 1, 2) #Beralih ke axes kedua dengan mengganti parameter ketiga dengan "2"
plt.plot(x2, y2, 'g.-')
plt.ylabel("Nilai $cos(x_2)$")
plt.xlabel("Nilai x")

plt.show()
```
