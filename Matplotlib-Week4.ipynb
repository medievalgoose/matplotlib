# Bar Plot / Bar Chart

## Import modules

```python
%matplotlib inline
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

## Simple Bar Plot

```python
data = [25, 45, 55, 125, 155]
kategori = ['A', 'B', 'C', 'D', 'E']
```

```python
plt.bar(kategori, data)
plt.grid()

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot')

plt.show()
```

Bar Plot setidaknya membutuhkan **dua** parameter, yaitu nilai untuk x dan nilai untuk y.

*Histogram digunakan untuk melakukan plotting data bertipe numerik dan yang sifatnya continuous, sedangkan bar plot digunakan untuk plotting data yang bersifat categorical dan discrete.* 

## Pengaturan Grid dan Color

```python
data = [25, 45, 55, 125, 155]
kategori = ['A', 'B', 'C', 'D', 'E']
```

```python
plt.bar(kategori,
       data,
       color='red',
       alpha=0.25)

plt.grid(linestyle='--',
        linewidth=1,
        axis='y',
        alpha=0.50)

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot')

plt.show()
```

Seperti yang sudah dicontohkan di atas, kita bisa mengatur baik warna dari bar plot dan ketebalan warnanya. Grid-nya juga bisa diatur dengan menggunakan `linestyle`, `linewidth`, `axis`, dan `alpha`.

## Horizontal Bar Plot

```python
data = [25, 45, 55, 125, 155]
kategori = ['A', 'B', 'C', 'D', 'E']
```

```python
plt.barh(kategori, data)
#plt.barh(kategori[::-1], data[::-1]) | Digunakan untuk membalik urutan.

plt.grid(linestyle='--',
        linewidth=1,
        axis='x',
        alpha=0.75)

plt.xlabel('Jumlah')
plt.ylabel('Kategori')
plt.title('Contoh Bar Plot')

plt.show()
```

Untuk membuat horizontal bar plot, dapat digunakan method `barh()`.

## Stacked Bar Plot

```python
data1 = [25, 85, 75, 40, 60]
data2 = [40, 35, 20, 55, 10]
kategori = ['A', 'B', 'C', 'D', 'E']
```

```python
plt.bar(kategori, data1, label='Data 1')
#plt.bar(kategori, data2, label='Data 2')
plt.bar(kategori, data2, label='Data 2', bottom=data1)

plt.legend()

plt.grid(linestyle='--',
        linewidth=1,
        axis='y',
        alpha=0.75)

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot')

plt.show()
```

* Terdapat parameter **bottom** untuk men-spesifikasi dimana data 2 akan ditempatkan.

* Jika tidak menspesifikasikan parameter bottom, maka hasilnya plottingnya akan tertimpa.

## Grouped Bar Plot

```python
data1 = [25, 85, 75, 40, 60]
data2 = [40, 35, 20, 55, 10]
kategori = ['A', 'B', 'C', 'D', 'E']

x = np.arange(len(kategori))
x
```

```python
width = 0.35

plt.bar(x-width/2, data1, width, label='Data 1')
plt.bar(x+width/2, data2, width, label='Data 2')

plt.xticks(x, kategori)

plt.legend()

plt.grid(linestyle='--',
        linewidth=1,
        axis='y',
        alpha=0.75)

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot')

plt.show()
```

Pada Grouped Bar Plot, hasil bar plot dari kedua data akan diplotting berdampingan. 

Proses plotting grouped bar plot:
* Menspesifikasikan `width` / lebar dari bar-nya.
* Menggunakan method `bar()`, namun parameter pertamanya berasosiasi dengan nilai index dari kategori. Parameter ketiga digunakan untuk merepresentasikan width-nya.
* Mengatur method `xticks()` untuk mengatur index.

# Pie Plot / Pie Chart

## Import modules

```python
%matplotlib inline
```

```python
import matplotlib
import matplotlib.pyplot as plt
```

## Simple Pie Plot

```python
kategori = ['A', 'B', 'C', 'D', 'E']
data = [215, 130, 245, 210, 225]
```

```python
plt.pie(data, 
        labels=kategori,
       autopct='%1.1f%%',
       startangle=90)

plt.title('Contoh Pie Plot')
plt.show()
```

Dilakukan dengan menggunakan fungsi `plt.pie()`. Parameter pertama akan berasosiasi dengan sejumlah integer yang nantinya akan dilakukan pembagian persentase untuk potongan pie-nya. Parameter berikutnya adalah `labels` yang akan berasosiasi dengan kategori.

Selanjutnya adalah `autopct`, yaitu untuk menampilkan persentase dari tiap potongan (bersifat opsional). Sedangankan parameter `startangle` dilakukan untuk melakukan rotasi pada pie plotnya.

## Pengaturan Legend dan Color

```python
kategori = ['A', 'B', 'C', 'D', 'E']
data = [215, 130, 245, 210, 225]
warna = ['pink', 'cyan', '#c03664', 'yellowgreen', 'skyblue']
```

```python
plt.pie(data,
       labels=kategori,
       colors=warna,
       autopct='%1.1f%%',
       startangle=90)

plt.legend()
plt.title('Contoh Pie Plot')

plt.show()
```

Terdapat juga parameter `colors` yang dapat diatur sesuai keinginan. Pada kasus ini, warna yang akan ditampilkan berasosiasi dengan list warna.

## Exploded Pie Plot

```python
kategori = ['A', 'B', 'C', 'D', 'E']
data = [215, 130, 245, 210, 225]
warna = ['pink', 'cyan', '#c03664', 'yellowgreen', 'skyblue']
explode_var = [0., 0., 0.2, 0., 0.08]
```

```python
plt.pie(data,
       labels=kategori,
       colors=warna,
        explode=explode_var,
       autopct='%1.1f%%',
       startangle=90)

plt.legend()
plt.title('Contoh Pie Plot')

plt.show()
```

Ada parameter `explode` yang dapat digunakan bagian potongan mana yang akan di-"explode". Pada kasus ini, nilai explodenya berkorelasi dengan variable explode_var.
