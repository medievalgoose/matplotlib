# Box Plot / Box Chart

## Import modules

```python
%matplotlib inline
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

## Sample Dataset

```python
np.random.seed(2)
data = np.random.normal(loc=100, scale=10, size=200)
data
```

Dengan menggunakan method `np.random.normal()`, ada beberapa parameter yang harus dispesifikasi. Pertama adalah parameter `loc` untuk mengatur nilai tengahnya. Parameter `scale` akan mengatur sebaran data / standard deviation, dan parameter `size`, yaitu nilai jumlah yang ingin dihasilkan.

## Simple Box Plot

```python
plt.boxplot(data, labels=['Data'])

plt.title('Simple Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.grid()

plt.show()
```

## Notched Box Plot

```python
plt.boxplot(data, labels=['Data'], notch=True)

plt.title('Simple Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.grid()

plt.show()
```

Dilakukan dengan menambahkan parameter `notch` yang diberi nilai `True`.

## Pengaturan marker untuk outlier

```python
plt.boxplot(data, labels=['Data'],
           showfliers=True, # Bisa diganti menjadi False jika tidak ingin menampilkan outlier.
           flierprops={'markerfacecolor':'r', 'marker':'d'}) # d = diamond.

plt.title('Simple Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.grid()

plt.show()
```

## Horizontal Box Plot

```python
plt.boxplot(data, labels=['Data'], vert=False) # Vert = vertical.

plt.title('Simple Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.grid()

plt.show()
```

## Multiple Box Plot

```python
np.random.seed(10)

data1 = np.random.normal(100, 10, 200)
data2 = np.random.normal(80, 30, 200)
data3 = np.random.normal(90, 20, 200)
data4 = np.random.normal(70, 25, 200)

data = [data1, data2, data3, data4]
labels = ['Data 1', 'Data 2', 'Data 3', 'Data 4']
```

```python
plt.boxplot(data, labels=labels)

plt.title('Simple Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.grid()

plt.show()
```

Garis tengah dari box plot merepresentasikan nilai **median**.

# Violin Plot / Violin Chart

## Import modules

```python
%matplotlib inline
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

## Sample Dataset

```python
np.random.seed(2)
data = np.random.normal(loc=100, scale=10, size=200)
data
```

* `loc` = nilai mean / rata-rata.
* `scale` = standard deviation / sebaran data.
* `size` = banyaknya data yang dihasilkan.

## Simple Violin Plot

```python
plt.violinplot(data)

plt.title('Simple Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

Garis yang terbentuk mengindikasikan nilai terbesar dan nilai terkecil dari dataset.

## Pengaturan pada Violin Plot

```python
plt.violinplot(data,
              showextrema=True,
              showmeans=False,
              showmedians=False,
              quantiles={0.25, 0.5, 0.75})

plt.title('Simple Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

* `showextrema` digunakan untuk menampilkan garis tengah, garis penanda nilai terbesar dan garis penanda nilai terkecil.

* `showmeans` digunakan untuk menampilkan nilai mean.

* `showmedian` digunakan untuk menampilkan nilai median.

* `quantiles` akan menerima sekumpulan data yang akan merepresentasikan persentase.

## Horizontal Violin Plot

```python
plt.violinplot(data, vert=False) # Vert = vertical.

plt.title('Simple Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

## Multiple Violin Plot

```python
np.random.seed(10)

data1 = np.random.normal(100, 10, 200)
data2 = np.random.normal(80, 30, 200)
data3 = np.random.normal(90, 20, 200)
data4 = np.random.normal(70, 25, 200)

data = [data1, data2, data3, data4]
```

```python
plt.violinplot(data)

plt.title('Simple Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```
