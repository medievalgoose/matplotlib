## Predefined Styles

Terdapat banyak predefined styles yang disediakan oleh Matplotlib. Informasi lebih lanjut bisa diakses [di sini](https://github.com/matplotlib/matplotlib/tree/main/lib/matplotlib/mpl-data/stylelib)

```python
plt.style.available
```

```python
plt.style.use('ggplot')
```

```python
x = np.linspace(1, 10, 15)

plt.plot(x, x, 'bs-', label='lin')
plt.plot(x, np.log(x), 'ro-', label='log')

plt.legend()
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Contoh Plot')

plt.show()
```

## Menerapkan multiple style

```python
plt.style.use(['seaborn', 'dark_background'])
```

Kode di atas berarti akan digunakan style seaborn dan akan ada beberapa aspek yang akan ditimpa dengan dark_background.

```python
x = np.linspace(1, 10, 15)

plt.plot(x, x, 'bs-', label='lin')
plt.plot(x, np.log(x), 'ro-', label='log')

plt.legend()
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Contoh Plot')

plt.show()
```

## Menerapkan temporary styling

```python
plt.style.use('ggplot')

plt.plot(np.sin(np.linspace(0, 2 * np.pi)), 'r-o')
plt.show()
```

```python
with plt.style.context('dark_background'):
    plt.plot(np.sin(np.linspace(0, 2* np.pi)), 'r-o')
    
plt.show()
```

Untuk melakukan temporary styling, dapat digunakan klausa `with` yang diikuti dengan `plt.style.context` lalu menspesifikasikan style yang ingin digunakan.

```python
plt.plot(np.sin(np.linspace(0, 2 * np.pi)), 'r-o')
plt.show()
```
