## Import modules

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

## Membuat plotting sederhana

```python
plt.plot([2, 5, 7, 11]) #Mulai melakukan plotting.
plt.ylabel("sumbu y") #Pemberian label pada sumbu Y.
plt.show() #Jika tidak menggunakan plt.show(), akan ada teks tambahan.
```

* Pemanggilan fungsi plot() dapat dilakukan dengan menyertakan hanya sebuah deret bilangan (list / array) ataupun dua buah deret bilangan (list / array).

* Bila pemanggilan hanya menyertakan sebuah deret bilangan, maka nilai pada deret bilangan tersebut akan dijadikan data point pada **sumbu y**; sedangkan nilai pada sumbu x akan secara otomatis dibuatkan sebuah deret bilangan terutu dengan nilai dimulai dari **nol**.

* Pada kasus di atas, saat nilai sumbu Y-nya 2, maka nilai x nya 0. Saat nilai y 5, maka nilai x nya 1, dan seterusnya.

```python
plt.plot([2, 3, 4, 7], [5, 4, 2, 16])
plt.show()
```

* Dengan kode di atas, deret bilangan pertama akan menjadi nilai sumbu x, dan deretan kedua akan menjadi nilai sumbu y.

## Pengaturan Format pada Plot

* Selain menyertakan dua buah deret bilangan sebagai paramter untuk sumbu x dan y, terdapat parameter ketiga yang bisa kita sertakan untuk mengatur warna dan jenis plotting. 

* Pengaturan parameter ketiga ini sepenuhnya mengadopsi gaya formatting pada MATLAB. 

* Nilai default untuk parameter ketiga ini adalah 'b-'. b berarti blue, dan - (dash) berarti garis.

* Pemanggilan fungsi axis() menyertakan list [xmin, xmax, ymin, ymax]

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro') #'ro' berarti merah dan default markernya berupa dot.
plt.axis([0, 6, 0, 20]) #Penentuan nilai sumbu x min dan max, serta nilai y min dan max.
plt.show()
```
