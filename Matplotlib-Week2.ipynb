## Import modules

```python
%matplotlib inline
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

## Membuat plotting sederhana

```python
plt.plot([2, 5, 7, 11]) #Mulai melakukan plotting.
plt.ylabel("sumbu y") #Pemberian label pada sumbu Y.
plt.show() #Jika tidak menggunakan plt.show(), akan ada teks tambahan.
```

* Pemanggilan fungsi plot() dapat dilakukan dengan menyertakan hanya sebuah deret bilangan (list / array) ataupun dua buah deret bilangan (list / array).

* Bila pemanggilan hanya menyertakan sebuah deret bilangan, maka nilai pada deret bilangan tersebut akan dijadikan data point pada **sumbu y**; sedangkan nilai pada sumbu x akan secara otomatis dibuatkan sebuah deret bilangan terutu dengan nilai dimulai dari **nol**.

* Pada kasus di atas, saat nilai sumbu Y-nya 2, maka nilai x nya 0. Saat nilai y 5, maka nilai x nya 1, dan seterusnya.

```python
plt.plot([2, 3, 4, 7], [5, 4, 2, 16])
plt.show()
```

* Dengan kode di atas, deret bilangan pertama akan menjadi nilai sumbu x, dan deretan kedua akan menjadi nilai sumbu y.

## Pengaturan Format pada Plot

* Selain menyertakan dua buah deret bilangan sebagai paramter untuk sumbu x dan y, terdapat parameter ketiga yang bisa kita sertakan untuk mengatur warna dan jenis plotting. 

* Pengaturan parameter ketiga ini sepenuhnya mengadopsi gaya formatting pada MATLAB. 

* Nilai default untuk parameter ketiga ini adalah 'b-'. b berarti blue, dan - (dash) berarti garis.

* Pemanggilan fungsi axis() menyertakan list [xmin, xmax, ymin, ymax]

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro') #'ro' berarti merah dan default markernya berupa dot.
plt.axis([0, 6, 0, 20]) #Penentuan nilai sumbu x min dan max, serta nilai y min dan max.
plt.show()
```

## Multi plot dalam satu pemanggilan fungsi plot()

```python
t = np.arange(0., 5., 0.2)
t
```

Fungsi arange() membutuhkan tiga parameter; 1. nilai min, 2. nilai max, 3. interval.

Hal yang membedakan linspace dan arange adalah parameter ketiganya. Untuk linspace, parameter ketiga berfungsi untuk mengatur jumlah data point yang akan berada diantara nilai min dan max nya, sedangkan pada arange, parameter ketiga untuk mengatur interval.

```python
plt.plot(t, t, 'r--') #Hasil plotting akan berupa garis putus-putus berwarna merah.
plt.plot(t, t**2, 'bs') #Hasil plotting berupa square dengan warna biru.
plt.plot(t, t**3, 'g*') #Nilai hasil plotting diambil dari nilai y pangkat tiga, dan markernya berupa bintang berwarna hijau.
plt.show()
```

Kita juga dapat melakukan multi plot dalam satu pemanggilan fungsi plot()

```python
plt.plot(t, t, 'r--',
        t, t**2, 'bs',
        t, t**3, 'g*') #Tiap plotting dipisahkan dengan tanda koma.
plt.show()
```

## Plotting dengan keywords

Matplotlib juga memungkinkan kita untuk melakukan plotting sekumpulan nilai yang tersimpan dalam struktur data yang disertai keywords argument (kwargs) seperti pada dictionary dan Pandas DataFrame.

```python
data = {'a': np.arange(50),
       'c': np.random.randint(0, 50, 50),
       'd': np.random.randn(50)}

data['b'] = data['a'] + 10 * np.random.randn(50) #Nilai key B akan mengadopsi nilai dari key A.
data['d'] = np.abs(data['d']) * 100

data
```

```python
plt.scatter('a', 'b', data = data) #Plotting dengan scatter plot.
plt.xlabel("entry a")
plt.ylabel("entry b")
plt.show()
```

Karena datanya berbentuk dictionary, maka nilai x dan y nya diambil dari key yang ada di dictionary tersebut. Pada kasus ini, key A dan key B.

Scatter plot juga bisa menampilkan dimensi ketiga dan keempat. 

```python
plt.scatter('a', 'b', c='c', s = 'd', data = data)
plt.xlabel("entry a")
plt.ylabel("entry b")
plt.show()
```

Parameter ketiga akan merepresentasikan warna, sedangkan parameter ke empat akan merepresentasikan size.

## Plotting untuk tipe data categorical

Matplotlib juga mendukung plotting untuk data dengan tipe kategori atau biasa dikenal dengan istilah categorical data atau categorical variables.

```python
names = ['group a', 'group b', 'group c']
values = [1, 10, 100]
```

```python
plt.plot(names, values)
plt.show()
```

```python
plt.scatter(names, values)
plt.show()
```

```python
plt.bar(names, values)
plt.show()
```

* Nilai group a akan berkorelasi dengan nilai 1, nilai group b akan berkorelasi dengan nilai 10, dan niali group c akan berkorelasi dengan nilai 100.
